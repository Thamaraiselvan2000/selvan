# 1. Write a python program to create a base class "Shape" with methods to calculate area and perimeter. Then, create derived classes "Circle" and "Rectangle" that inherit from the base class and calculate their respective areas and perimeters. Demonstrate their usage in a program.

You are developing an online quiz application where users can take quizzes on

various topics and receive scores.

1. Create a class for quizzes and questions. 2. Implement a scoring system that calculates the user's score on a quiz.

3. How would you store and retrieve user progress, including quiz history and

scores?

```python
import math

class Shape:
    def calculate_area(self):
        pass
    
    def calculate_perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def calculate_area(self):
        return math.pi * self.radius**2
    
    def calculate_perimeter(self):
        return 2 * math.pi * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def calculate_area(self):
        return self.length * self.width
    
    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

# Usage example
circle = Circle(5)
print(f"Circle Area: {circle.calculate_area()}")
print(f"Circle Perimeter: {circle.calculate_perimeter()}")

rectangle = Rectangle(4, 6)
print(f"Rectangle Area: {rectangle.calculate_area()}")
print(f"Rectangle Perimeter: {rectangle.calculate_perimeter()}")
```

```python
class Question:
    def __init__(self, text, options, correct_option):
        self.text = text
        self.options = options
        self.correct_option = correct_option

    def is_correct(self, selected_option):
        return selected_option == self.correct_option

class Quiz:
    def __init__(self, name):
        self.name = name
        self.questions = []

    def add_question(self, question):
        self.questions.append(question)

    def take_quiz(self):
        score = 0
        total_questions = len(self.questions)
        for question in self.questions:
            print(question.text)
            for i, option in enumerate(question.options, start=1):
                print(f"{i}. {option}")
            selected_option = int(input("Enter the number of your answer: "))
            if question.is_correct(selected_option):
                score += 1
                print("Correct!\n")
            else:
                print("Incorrect.\n")
        return score, total_questions

# Example 
q1 = Question("What is the capital of India?", ["Chennai", "Delhi", "Mumbai", "kolkata"], 1)
q2 = Question("which is the largest planet in the solar system?", ["Earth", "Mars", "Venus", "Jupiter"], 4)

quiz = Quiz("General Knowledge Quiz")
quiz.add_question(q1)
quiz.add_question(q2)

user_score, total_questions = quiz.take_quiz()
print(f"Your Score: {user_score}/{total_questions}")
```

```python
import json

class User:
    def __init__(self, username):
        self.username = username
        self.quiz_history = {}

    def take_quiz(self, quiz, score):
        self.quiz_history[quiz.name] = score

    def get_quiz_history(self):
        return self.quiz_history

# Example 
user1 = User("Rohit")
user2 = User("John")

quiz1 = Quiz("General Knowledge Quiz")
quiz2 = Quiz("Science Quiz")

user1.take_quiz(quiz1, 2)
user1.take_quiz(quiz2, 1)

user2.take_quiz(quiz1, 3)
user2.take_quiz(quiz2, 2)

# Storing user progress to a file (JSON)
users = [user1, user2]
with open("user_progress.json", "w") as file:
    user_data = [{"username": user.username, "quiz_history": user.quiz_history} for user in users]
    json.dump(user_data, file)

# Retrieving user progress from the file
with open("user_progress.json", "r") as file:
    user_data = json.load(file)
    for user_info in user_data:
        username = user_info["username"]
        quiz_history = user_info["quiz_history"]
        print(f"{username}'s Quiz History:")
        for quiz_name, score in quiz_history.items():
            print(f"{quiz_name}: {score}")
```

# 3. You are tasked with designing a Python class hierarchy for a simple banking system. The system should be able to handle different types of accounts, such as Savings Accounts and Checking Accounts. Both account types should have common attributes like an account number, account holder's name, and balance. However, Savings Accounts should have an additional attribute for interest rate, while Checking Accounts should have an attribute for overdraft limit.

1. Create a Python class called BankAccount with the following attributes and methods:

a. Attributes: account number, account holder name, balance b. Methods: init_0) (constructor), deposit(), and withdraw()

2. Create two subclasses, Savings Account and Checking Account, that inherit from the BankAccount class,

3. Add the following attributes and methods to each subclass:

a Savings Account:

i. Additional attribute: interest rate ii. Method: calculate interest(), which calculates and adds interest

to the account based on the interest rate.

b. Checking Account:

1. Additional attribute: overdraft limit

ii. Method: withdraw(), which allows withdrawing money up to the overdraft limit (if available) without additional fees.

4. Write a program that creates instances of both Savings Account and

Checking Account and demonstrates the use of their methods.

5. Implement proper encapsulation by making the attributes private where

necessary and providing getter and setter methods as needed. 6. Handle any potential errors or exceptions that may occur during operations

like withdrawals, deposits, or interest calculations.

7. Provide comments in your code to explain the purpose of each class, attribute,

and method.

Note: Your code should create instances of the classes, simulate transactions, and

showcase the differences between Savings Accounts and Checking Accounts

```python
class BankAccount:
    """
    Base class for a generic bank account.
    Attributes:
    - account_number (str): The account number.
    - account_holder_name (str): The account holder's name.
    - _balance (float): The account balance (private attribute).
    Methods:
    - __init__(self, account_number, account_holder_name, initial_balance): Constructor to initialize account details.
    - deposit(self, amount): Deposit money into the account.
    - withdraw(self, amount): Withdraw money from the account.
    - get_balance(self): Get the account balance.
    """
    def __init__(self, account_number, account_holder_name, initial_balance):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self._balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if 0 < amount <= self._balance:
            self._balance -= amount
        else:
            print("Insufficient funds for withdrawal.")

    def get_balance(self):
        return self._balance


class SavingsAccount(BankAccount):
    """
    Subclass representing a savings account.
    Additional Attributes:
    - interest_rate (float): The interest rate.
    Methods:
    - __init__(self, account_number, account_holder_name, initial_balance, interest_rate):
    Constructor to initialize savings account details.
    - calculate_interest(self): Calculate and add interest to the account based on the interest rate.
    """
    def __init__(self, account_number, account_holder_name, initial_balance, interest_rate):
        super().__init__(account_number, account_holder_name, initial_balance)
        self.interest_rate = interest_rate

    def calculate_interest(self):
        interest = self.get_balance() * (self.interest_rate / 100)
        self.deposit(interest)


class CheckingAccount(BankAccount):
    """
    Subclass representing a checking account.
    Additional Attribute:
    - overdraft_limit (float): The overdraft limit.
    Methods:
    - __init__(self, account_number, account_holder_name, initial_balance, overdraft_limit):
    Constructor to initialize checking account details.
    - withdraw(self, amount): Override withdraw method to allow overdraft up to the limit.
    """
    def __init__(self, account_number, account_holder_name, initial_balance, overdraft_limit):
        super().__init__(account_number, account_holder_name, initial_balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if 0 < amount <= (self.get_balance() + self.overdraft_limit):
            self._balance -= amount
        else:
            print("Withdrawal exceeds overdraft limit.")


# Example 
if __name__ == "__main__":
    # Create a Savings Account
    savings_acc = SavingsAccount("SA123", "Rohit", 1000.0, 3.5)
    savings_acc.calculate_interest()
    print(f"Savings Account Balance: ${savings_acc.get_balance()}")

    # Create a Checking Account
    checking_acc = CheckingAccount("CA456", "John", 500.0, 100.0)
    checking_acc.withdraw(700.0)
    print(f"Checking Account Balance: ${checking_acc.get_balance()}")
```

# 5. Library Management System-Scenario: You are developing a library management system where you need to handle books, patrons, and library transactions

1. Create a class hierarchy that includes classes for books (eg, Book), patrons (e.g...

Patron), and transactions (e.g., Transaction). Define attributes and methods for each class. 2. Implement encapsulation by making relevant attributes private and providing getter

and setter methods where necessary.

3. Use inheritance to represent different types of books (e.g.. fiction, non-fiction) as subclasses of the Book class. Ensure that each book type can have specific attributes

and methods.

4. Demonstrate polymorphism by allowing patrons to check out and return books..

regardless of the book type.

5. Implement a method for tracking overdue books and notifying patrons. 6. Consider scenarios like book reservations, late fees, and library staff interactions in your design.

```python
class Book:
    def __init__(self, title, author, book_type, isbn, available=True):
        self._title = title
        self._author = author
        self._book_type = book_type
        self._isbn = isbn
        self._available = available

    def get_title(self):
        return self._title

    def set_title(self, title):
        self._title = title

    def get_author(self):
        return self._author

    def set_author(self, author):
        self._author = author

    def get_book_type(self):
        return self._book_type

    def set_book_type(self, book_type):
        self._book_type = book_type

    def get_isbn(self):
        return self._isbn

    def set_isbn(self, isbn):
        self._isbn = isbn

    def is_available(self):
        return self._available

    def check_out(self):
        if self._available:
            self._available = False
            return True
        else:
            return False

    def return_book(self):
        if not self._available:
            self._available = True
            return True
        else:
            return False

class FictionBook(Book):
    def __init__(self, title, author, isbn, available=True, genre=None):
        super().__init__(title, author, "Fiction", isbn, available)
        self._genre = genre

    def get_genre(self):
        return self._genre

    def set_genre(self, genre):
        self._genre = genre

class NonFictionBook(Book):
    def __init__(self, title, author, isbn, available=True, subject=None):
        super().__init__(title, author, "Non-Fiction", isbn, available)
        self._subject = subject

    def get_subject(self):
        return self._subject

    def set_subject(self, subject):
        self._subject = subject

class Patron:
    def __init__(self, name, patron_id):
        self._name = name
        self._patron_id = patron_id

    def get_name(self):
        return self._name

    def set_name(self, name):
        self._name = name

    def get_patron_id(self):
        return self._patron_id

    def set_patron_id(self, patron_id):
        self._patron_id = patron_id

    def check_out_book(self, book):
        if book.check_out():
            return True
        else:
            return False

    def return_book(self, book):
        if book.return_book():
            return True
        else:
            return False

class Transaction:
    def __init__(self, book, patron, due_date):
        self._book = book
        self._patron = patron
        self._due_date = due_date

    def get_book(self):
        return self._book

    def get_patron(self):
        return self._patron

    def get_due_date(self):
        return self._due_date

def check_overdue_transactions(transactions, current_date):
    overdue_transactions = []
    for transaction in transactions:
        if current_date > transaction.get_due_date():
            overdue_transactions.append(transaction)
    return overdue_transactions

# Example 
if __name__ == "__main__":
    fiction_book = FictionBook("The Great ", "F. Scott Fitzgerald", "978-0743273565", genre="Classic")
    non_fiction_book = NonFictionBook("The Glass Castle", "Yuval Noah Harari", "978-0062316097", subject="History")
    patron = Patron("Rohit", "P001")

    transaction1 = Transaction(fiction_book, patron, "2023-09-30")
    transaction2 = Transaction(non_fiction_book, patron, "2023-09-15")

    transactions = [transaction1, transaction2]

    current_date = "2023-09-20"

    overdue_transactions = check_overdue_transactions(transactions, current_date)

    if overdue_transactions:
        print("Overdue Transactions:")
        for transaction in overdue_transactions:
            print(f"Book: {transaction.get_book().get_title()}, Due Date: {transaction.get_due_date()}")

    print("Before checkout:", fiction_book.is_available())
    patron.check_out_book(fiction_book)
    print("After checkout:", fiction_book.is_available())
    patron.return_book(fiction_book)
    print("After return:", fiction_book.is_available())
```

# 6.Online Shopping Cart

Scenario: You are tasked with designing a class hierarchy for an online shopping cart system. The system should handle products, shopping carts, and orders. Consider various OOP principles while designing this system.

1. Create a class hierarchy that includes classes for products (e.g., Product), shopping carts (eg. ShoppingCart), and orders (e.g., Order). Define attributes and methods for each class.

2. Implement encapsulation by making relevant attributes private and providing getter and

setter methods where necessary

3. Use inheritance to represent different types of products (e.g., electronics, clothing) as subclasses of the Product class. Ensure that each product type can have specific attributes

and methods.

4 Demonstrate polymorphism by allowing various product types to be added to a shopping cart and calculate the total cost of items in the cart. 5. Implement a method for placing an order, which transfers items from the shopping cart to an order.

Consider scenarios like out-of-stock products, discounts, and shipping costs in your design.

```python
class Product:
    def __init__(self, product_id, name, price):
        self._product_id = product_id
        self._name = name
        self._price = price

    def get_product_id(self):
        return self._product_id

    def set_product_id(self, product_id):
        self._product_id = product_id

    def get_name(self):
        return self._name

    def set_name(self, name):
        self._name = name

    def get_price(self):
        return self._price

    def set_price(self, price):
        self._price = price

class ElectronicsProduct(Product):
    def __init__(self, product_id, name, price, brand):
        super().__init__(product_id, name, price)
        self._brand = brand

    def get_brand(self):
        return self._brand

    def set_brand(self, brand):
        self._brand = brand

class ClothingProduct(Product):
    def __init__(self, product_id, name, price, size):
        super().__init__(product_id, name, price)
        self._size = size

    def get_size(self):
        return self._size

    def set_size(self, size):
        self._size = size

class ShoppingCart:
    def __init__(self):
        self._items = []

    def add_item(self, product, quantity=1):
        self._items.append({'product': product, 'quantity': quantity})

    def remove_item(self, product):
        self._items = [item for item in self._items if item['product'] != product]

    def calculate_total(self):
        total = 0
        for item in self._items:
            total += item['product'].get_price() * item['quantity']
        return total

class Order:
    def __init__(self, order_id, cart, shipping_cost=0, discount=0):
        self._order_id = order_id
        self._cart = cart
        self._shipping_cost = shipping_cost
        self._discount = discount

    def get_order_id(self):
        return self._order_id

    def set_order_id(self, order_id):
        self._order_id = order_id

    def place_order(self):
        total_cost = self._cart.calculate_total() + self._shipping_cost - self._discount
        if total_cost < 0:
            total_cost = 0
        return total_cost

# Example
if __name__ == "__main__":
    laptop = ElectronicsProduct("P001", "Laptop", 800, "HP")
    tshirt = ClothingProduct("P002", "T-Shirt", 20, "M")
    phone = ElectronicsProduct("P003", "Smartphone", 400, "Samsung")

    cart = ShoppingCart()
    cart.add_item(laptop, 2)
    cart.add_item(tshirt, 3)
    cart.add_item(phone)

    order = Order("O001", cart, shipping_cost=10, discount=50)
    total_cost = order.place_order()

    print("Order ID:", order.get_order_id())
    print("Total Cost:", total_cost)
```
